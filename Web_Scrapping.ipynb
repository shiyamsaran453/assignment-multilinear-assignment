# Web Scrapping

```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
import requests
```

```python
Url="https://www.vijaysales.com/mobiles-and-tablets/type/buy-smartphones-mobile-phone"
```

```python
response=requests.get(Url)
```

```python
response.raise_for_status()
```

```python
Back_code=BeautifulSoup(response.text)
```

```python
name=Back_code.select('.BcktPrdNm_')
```

```python
price=Back_code.select('.Prdvsprc_')
```

```python
offer=Back_code.select('.Prdmrprc_')
```

```python
Offer_per=Back_code.select('.ofrnm_')
```

```python
Mobile_data=pd.DataFrame(name,columns=['Name'])
```

```python
Mobile_data["Price"]=price
```

```python
Mobile_data["Offer_price"]=offer
```

```python
Mobile_data["Percentage_Offer"]=Offer_per
```

```python
Mobile_data
```

```python
Mobile_data.to_csv('Mobile_data.csv')
```
